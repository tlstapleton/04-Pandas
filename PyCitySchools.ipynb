```python
grouped_df_3=school_data_complete.groupby(['school_name','type','size','budget'])
math_df_2=pd.DataFrame(grouped_df_3['math_score','reading_score'].mean())
math_df_2.reset_index(inplace=True)
math_df_2['Per Student Budget']=math_df_2['budget']/math_df_2['size']
math_df_2
```

```python
math_pass_df=pd.DataFrame(school_data_complete[school_data_complete['math_score'] >=70].groupby('school_name')['math_score'].count())
rdg_pass_df=pd.DataFrame(school_data_complete[school_data_complete['reading_score'] >=70].groupby('school_name')['reading_score'].count())
passing_numbers=pd.merge(math_pass_df,rdg_pass_df,on='school_name')
math_df_2=pd.merge(math_df_2,passing_numbers,on='school_name',suffixes=('','_countpass'))
math_df_2
```

```python
math_df_2['% Passing Math']=(math_df_2['math_score_countpass']/math_df_2['size'])*100
math_df_2['% Passing Reading']=(math_df_2['reading_score_countpass']/math_df_2['size'])*100
math_df_2['% Overall Passing Rate']=((math_df_2['% Passing Math']+math_df_2['% Passing Reading'])/2)
math_df_2.drop(columns=['math_score_countpass','reading_score_countpass'],axis=1,inplace=True)
math_df_2.rename(columns={'school_name':'School','type':'School Type','size':'Total Students','budget':'Total School Budget','math_score':'Average Math Score','reading_score':'Average Reading Score'},inplace=True)
math_df_2=math_df_2[['School','School Type','Total Students','Total School Budget','Per Student Budget','Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing Rate']]
math_df_2.set_index('School',inplace=True)
math_df_2
```
